```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv', sep=',')
ratings.head()
```

```python
ratings_min_max = ratings.groupby('userId').agg([min, max]).loc[:, ['timestamp']]
ratings_min_max.head()
```

```python
ratings_min_max = ratings.groupby('userId').min().loc[:, ['timestamp']]
ratings_min_max.rename(columns={'timestamp' : 'min'}, inplace=True)
ratings_min_max['max'] = ratings.groupby('userId').max().loc[:, ['timestamp']]
ratings_min_max.head()
```

```python
ratings_count = ratings.groupby('userId').count().loc[:, ['movieId']]
ratings_count.rename(columns={'movieId' : 'count'}, inplace=True)
ratings_count.head()
```

```python
ratings_final = ratings_min_max.join(ratings_count, how='left')
ratings_final.head()
```

```python
ratings_final['lifetime'] = ratings_final['max'] - ratings_final['min']
ratings_final.head()
```

```python
ratings_final.loc[ratings_final['count'] > 100].mean()['lifetime']
```

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv', sep=',')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.head()
```

```python
rate = ratings.groupby('movieId').mean().round(1)[['rating']]
rate.head()
```

```python
mr = movies.merge(rate, left_on='movieId', right_on='movieId').set_index('movieId')
mr.head()
```

```python
def movie_rate(rating):
    if rating <= 2:
        return 'low'
    elif rating <= 4:
        return 'middle'
    else:
        return 'high'
```

```python
mr['class'] = mr['rating'].apply(movie_rate)
mr.head()
```

```python
mr.groupby('class').count().sort_values('rating', ascending=False)
```

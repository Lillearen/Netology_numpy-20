```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv', sep=',')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.head()
```

```python
rate = ratings.groupby('movieId').mean()[['rating']]
rate.head()
```

```python
mr = movies.merge(rate, left_on='movieId', right_on='movieId').set_index('movieId')
mr.head()
```

```python
def movie_rate(rating):
    if rating <= 2:
        return 'low rating'
    elif rating <= 4:
        return 'average rating'
    else:
        return 'high rating'
```

```python
mr['class'] = mr['rating'].apply(movie_rate)
mr.head(10)
```

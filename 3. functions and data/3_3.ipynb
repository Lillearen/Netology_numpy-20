```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv', sep=',')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.head()
```

```python
years = [i for i in range(1950, 2011)]
```

```python
def production_year (title):
    for y in years:
        if str(y) in title:
            return y
    return 1900
```

```python
movies['year'] = movies['title'].apply(production_year)
movies.head()
```

```python
rate = ratings.groupby('movieId').mean().round(1)[['rating']]
rate.head()
```

```python
mr = movies.merge(rate, left_on='movieId', right_on='movieId').set_index('movieId')
mr.head()
```

```python
mr.groupby('year').mean('rating').sort_values('rating', ascending=False)
```
